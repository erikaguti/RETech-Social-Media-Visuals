```python

''''
Let's make a Zoo

'''

#Creating lists for how different people say the same thing (these will be used to parse the text)
dogfinder = ['dog', 'dogs', 'Dog','Dogs', 'pup']
catfinder = ['cat', 'cats', 'Cat','Cats', 'kitty']
fishfinder = ['fish']

#creating seperate dfs for each animal where the entries in the master df contain the words we are looking for
dogs = zoo[zoo.Pets.str.contains('|'.join(dogfinder))]
cats = zoo[zoo.Pets.str.contains('|'.join(catfinder))]
fish = zoo[zoo.Pets.str.contains('|'.join(fishfinder))]
```

```python
'''
For the next 3 cells we take each animal df and extract their count of each animal
    
    This will be done in 3 steps:
       1. Where there is text representing numbers, convert that text into integers
       2. extract the integers from the text and add them to a list
       3. add up the integers in the list to create a count of the number of dogs, cats, and fish      
'''

#convert text to numbers
dogs.Pets = dogs.Pets.str.replace(r'one|One', '1')
#extract numbers and add to a list that we will then count
dogcount =  dogs.Pets.str.extract(r'(\d)')
#add numbers
dogcount = dogcount.to_numpy()
dogcount = [int(i) for i in dogcount]
dog = sum(dogcount)
print(dog)
```

```python
#convert to numbers
cats.Pets = cats.Pets.str.replace(r'one|One', '1')
#extract numbers
catcount =  cats.Pets.str.extract(r'(\d)')
#add numbers
catcount = catcount.to_numpy()
catcount = [int(i) for i in catcount]
cat = sum(catcount)
cat
```

```python
#convert to numbers
fish.Pets = fish.Pets.str.replace(r'one|One', '1')
#extract numbers
fishcount =  fish.Pets.str.extract(r'(\d)')
#add numbers
fishcount = fishcount.to_numpy()
fishcount = [int(i) for i in fishcount]
fishies = sum(fishcount)
fishies
```

```python
'''
I am using a python package I found at this link: 
https://pywaffle.readthedocs.io/en/latest/examples/plot_with_characters_or_icons.html

I will also need to use matplot lib to create the visualization
       
'''

import pywaffle
import matplotlib.pyplot as plt
from pywaffle import Waffle

#adding 1 to cat for the other cat
data = {str(dog)+" "+'Dogs':dog,str(cat+1)+" "+'Cats':cat+1,str(fishies)+" "+'Fish':fishies}

fig = plt.figure(
    FigureClass=Waffle, 
    rows=4, 
    values=data,
    icons = ['dog', 'cat', 'fish'], icon_size = 40,
    legend={'loc': 'lower left', 'bbox_to_anchor': (0, -0.4), 'ncol': len(data), 'framealpha': 0, 'fontsize':15},
    colors=["#1c4c56", "#0aa9be", "#999999"],
    icon_legend = True,
    title={'label': 'The Re Tech Zoo', 'loc': 'left','fontdict': {
            'fontsize': 16, 'color':'#3f3e3e'}},
    figsize = (8,5)
)

plt.show(fig)
```
